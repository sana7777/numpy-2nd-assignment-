```python
# Rotate Image 

degrees = 90
img = np.array(Image.open('pp.jpg'))
img = img.sum(2) / (255*3) # converting to grayscale

fig = plt.figure(figsize=(8, 8))

fig.add_subplot(2, 9, 10)
plt.imshow(img)
plt.title("original")
img0 = img.copy()

# Algo: image(ndarray) -> transpose -> mirror image across y axis (middle column)
for i in range(degrees // 90):
    img0 = img0.T
    
    for i in range(img0.shape[0]):
        for j in range(0, img0.shape[1] // 2):
            
            c = img0[i, j]
            img0[i, j] = img0[i, img0.shape[1]-j-1]
            img0[i, img0.shape[1]-j-1] = c
        
fig.add_subplot(2, 2, 2)
plt.imshow(img0)
plt.title("rotated")
```

```python
# or you could have used np.rot90(img)
plt.imshow(np.rot90(img))
```

```python
# Nagitive of an Image
img = np.array(Image.open('pp.jpg'))
img_grey = img.sum(2) / (255*3) # summing over axis=2 (channel axis) to get grey scaled image

fig = plt.figure(figsize=(10,10))
img_grey = 255*3 - img_grey        # 255 * 3 because we added along channel axis previously
fig.add_subplot(1, 2, 1)
plt.imshow(img_grey)
plt.title('Negative of Grey image')

img = 255 - img
fig.add_subplot(1, 2, 2)
plt.imshow(img)
plt.title('Negative of RGB image')
```

```python
# Padding Black Spaces

img = np.array(Image.open('pp.jpg'))
img_grey = img.sum(2) / (255*3)
img0 = img_grey.copy()
img0 = np.pad(img0, ((100,100),(100,100)), mode='constant')
plt.imshow(img0)
```

```python
# Viualizing RGB Channel

img = np.array(Image.open("pp.jpg"))
img_R,img_G,img_B = img.copy(), img.copy(), img.copy()

img_R[:, :, (1, 2)] = 0
img_G[:, :, (0, 2)] = 0
img_B[:, :, (0, 1)] = 0

img_rgb = np.concatenate((img_R,img_G,img_B),axis=1)
plt.figure(figsize=(10,10))
plt.imshow(img_rgb)
```

```python
# Color Reduction
img = np.array(Image.open("pp.jpg"))
# Making Pixel values discrete by first division by // which gives int and then multiply by same factor
img0 = (img // 16) *16
img1 = (img // 64) *250
img2 = (img // 120) * 120
img3 = (img // 256) * 256
img_all = np.concatenate((img,img0,img1,img2,img3),axis=1)
plt.figure(figsize=(15,15))
plt.imshow(img_all)
```

```python

# Trim Image by Array Slicing
img = np.array(Image.open("pp.jpg"))
fig = plt.figure(figsize=(10,10))
fig.add_subplot(1, 2, 1)
plt.imshow(img)
plt.title("Orginal")
img0 = img[80:-180, 128:-108,:] # trimed method
fig.add_subplot(1,2,2)
plt.imshow(img0)
plt.title("Trimed")
```

```python
# Pasting with slice
src = np.array(Image.open('pp.jpg').resize((128, 128)))
dst = np.array(Image.open('pp.jpg').resize((256, 256))) // 4

dst_copy = dst.copy()
dst_copy[64:128, 128:192] = src[32:96, 32:96]

fig = plt.figure(figsize=(10, 10))

fig.add_subplot(1, 2, 1)
plt.imshow(src)
plt.title('Original')

fig.add_subplot(1, 2, 2)
plt.imshow(dst_copy)
plt.title('Pasted with slice')
```

```python
# Transpose anti clock wise

fig = plt.figure(figsize=(10, 10))

img = np.array(Image.open('pp.jpg'))
fig.add_subplot(1, 2, 1)
plt.imshow(img)
plt.title('Original')

img0 = img.transpose(1, 0, 2)

fig.add_subplot(1, 2, 2)
plt.imshow(img0)
plt.title('Flip_rotated')
```
